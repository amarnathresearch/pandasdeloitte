```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'id':[1,2,3,4,5],
                   'col1':[10,20,30,40,50],
                   'col2':[6,7, 8, 9, 10],
                   'col3':['red','pink','blue','green','yellow']})
```

```python
df1
```

```python
df2 = pd.DataFrame({'id':[1,5, 6],
                   'col1':[60,70,80],
                   'col2':[9, 10, 20],
                   'col3':['red','yellow', 'orange']})
```

```python
df2
```

```python
inner_join = pd.merge(df1, df2, on=['id'])
```

```python
inner_join
```

```python
inner_join = pd.merge(df1, df2, on=['id','col3'])
```

```python
inner_join
```

```python
inner_join = pd.merge(df1, df2, suffixes=['_left','_right'], on=['id','col3'])
```

```python
inner_join
```

```python
inner_join = pd.merge(df1, df2, suffixes=['_left','_right'], left_on='col3', right_on='col3')
```

```python
inner_join
```

```python
inner_join = pd.merge(df1, df2, suffixes=['_left','_right'], on=['id'])
```

```python
inner_join
```

```python
left_join = pd.merge(df1, df2, suffixes=['_left','_right'], on=['id'], how="left")
```

```python
left_join
```

```python
right_join = pd.merge(df1, df2, suffixes=['_left','_right'], on=['id'], how="right")
```

```python
right_join
```

```python
outer_join = pd.merge(df1, df2, suffixes=['_left','_right'], on=['id'], how="outer")
```

```python
outer_join
```

```python
df1
```

```python
df2
```

# Join Function


```python
df1.set_index('id').join(df2.set_index('id'),  lsuffix='_left', rsuffix="_right")
```

```python
df1.set_index(['id','col3']).join(df2.set_index(['id','col3']),  lsuffix='_left', rsuffix="_right")
```

```python
df1.join(df2, on='id', how="inner", lsuffix='_left', rsuffix="_right")
```

```python
df1.join(df2, on='id', how="right", lsuffix='_left', rsuffix="_right")
```

# Concatenation

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, join="inner")
```
